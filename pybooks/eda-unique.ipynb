```python
train_df = pd.read_csv('../data/train_dataset_train.csv')
y_train_df = train_df[['label']]
```

```python
ctrain_df = train_df.copy().drop(columns=['id','label'])
ctrain_df[['oper_type','oper_attr']] = ctrain_df['oper_type + oper_attr'].str.split('_', expand=True)
ctrain_df['hashes'] = ctrain_df.apply(lambda x: hash(tuple(x)), axis=1)
ctrain_df.sample(10)
```

```python
ctrain_df.shape[0]-ctrain_df['hashes'].unique().shape[0]
```

```python
train_no_package = ctrain_df.drop(columns = package_related)
train_no_package['hashes_no_pack'] = train_no_package.drop(columns='hashes').apply(lambda x: hash(tuple(x)),axis=1)
train_no_package['hashes_no_pack'].unique().shape[0]
```

```python
from tqdm import tqdm

uh, uhi = np.unique(train_no_package['hashes_no_pack'], return_index=True)

strangers = []
goods = []
uhashes = uh.copy()
np.random.shuffle(uhashes)
counter = 0
t = tqdm(uhashes)
for uhnp in t:
    mn = y_train_df[train_no_package['hashes_no_pack']==uhnp]['label'].mean()
    if mn not in [0.0, 1.0, 0, 1]:
#         print(uhnp, mn)
        strangers.append([uhnp,mn])
    else:
        goods.append([uhnp, mn])
#     counter += 1
    t.set_description(f"Goods: {len(goods)}, bads: {len(strangers)}")
```

```python
train_unique_selected = train_no_package.iloc[np.where(np.isin(train_no_package['hashes_no_pack'], [g[0] for g in goods]))]
train_unique_selected.head()
    
```
